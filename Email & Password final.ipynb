# Description:
> The project involves the development of a Python application focusing on email and password validation, alongside storing user data in a JSON file format, including a one-time password (OTP) in a dictionary structure.

# Key Components:

- > Importing necessary libraries for functionality.
- > Initialization of an empty dictionary to manage user details such as email, password, and OTP.
- > Implementation of various functions and methods catering to distinct operations.
- > Adaptation of these functions to acquire user input and persist the data in JSON format through file handling, facilitating retrieval as needed.

# Procedure:
- > Prompting the user for login credentials.
- > Continuation to password and OTP verification upon successful user data matching.
- > Providing an option for users to create a new account if no matching data is found.
- > During the signup process, ensuring the correct format of email and password according to predefined criteria.
- **Email Conditions:**
- Validating domain existence.
- Enforcing a single '@' symbol usage.
- Restricting special characters.
- Prohibiting emails starting with a digit.
- **Password Conditions:**
 -  Enforcing a length requirement between 8 to 16 characters.
 -  Requiring at least one uppercase letter.
 -  Mandating at least one lowercase letter.
 -  Ensuring the presence of at least one digit.
-  Requiring at least one special character.
- > Post-signup, offering the user the option to proceed with login.
- > Validating OTP and password authenticity during the login process.







```python
# Importing required libararies
from random import randrange
import json
```

```python
# creating empty dictionary 
dictionary = {}
# predefined domains
domain = ['gmail.com','yahoo.com','microsoft.com']
```

```python
# Defining a function for Login procedure
def login(): 
    print("LOGIN :")
    mail = input('Enter Email Address : ')
    if mail in dictionary.keys() :
        lpassword = input("Enter Login password : ")

        if lpassword == dictionary[mail][0] :
            print('check')
            login_otp(mail)
        else :
            print('Incorrect password for : ',mail)
    else :
        print('Entered email does not exist')
        inp = input('Want to signup (y/n) : ')
        if inp.lower() == 'y' :
            signup()
        else :
            print('Thankyou')

```

```python
# defining a function for Sign-up process
def signup():  
    
    email = input("Enter New email address : ")
    count = 0
    recicount = len(email)
    for i in email :
        if email[0].isdigit() :
            print('Email cannot start with a number')
            break

        if not i.isalnum() and (i != '@') and (i != '.'):
            print('No special characters allowed other than @')
            break
        if i == '@':
            count += 1

        elif i != '@' :
            recicount -= 1

    if count != 1  or recicount != 1 :
        print('Email should contain single @ ')      
    if email.split('@')[-1] not in domain :
        print('Invalid domain')

    else :
        for i in dictionary.keys():
            if i == email :
                print('email already exist')
            
        else:
            key = email
            pwd = input("Enter New Password : ")
            spl = ['!','@','#','$','%','^','&','*']

            lower = False
            upper = False
            num   = False
            char  = False


            if 8 <= len(pwd) < 16:

                for i in pwd :

                    if i.islower() :
                        lower = True
                    elif i.isupper() :
                        upper = True
                    elif i.isdigit() :
                        num = True 
                    elif i in spl :
                        char = True

            else :
                print('password should be between 8 - 16 characters long!')

            if upper and lower and num and char :
                password = pwd
                dictionary.update({key:[password,'Nan']})

                signup_otp(key)
            else :
                print('Thank you')
```

```python
# Function for Generating OTP while Signing up
def signup_otp(mail) :
    otp = randrange(100000,1000000)
    n = True
    for i in range(1,4):
                
                dictionary[mail][-1] = otp
                file = open('otp_file_new.json','w')
                json.dump(dictionary,file)
                file.close()


                otp_validation = int(input('Enter OTP to confirm Signup : '))
                if otp_validation == dictionary[mail][-1] :
                    
                    print('Account created successfully!')
                    login_consent = input('Continue with login? (y/n) : ')
                    if login_consent.lower() == 'y' :
                        login()
                        break
                    else :
                        print('Thank you')
                    
                else :
                    
                    print('Incorrect OTP!')
                    print(f'{i}/3 attempts')
                    n = n + 1
                    if n == 3 :
                        del dictionary[mail]
                        file = open('otp_file_new.json','w')
                        json.dump(dictionary,file)
                        file.close()
                        print('Sign up failed Try again')
                        
                    
```

```python
# Function for generating OTP for Login Process
def login_otp(mail):
    otp = randrange(100000,1000000)
    for i in range(1,4) :
        dictionary[mail][-1] = otp
        file = open('otp_file_new.json','w')
        json.dump(dictionary,file)
        file.close()

        otp_validation = int(input('Enter OTP to Login : '))
        if otp_validation == dictionary[mail][-1] :
            print('Login Successful!')
            break
        else :
            print('Incorrect OTP!')
            print(f'{i}/3 attempts')
```
